```python
import csv

names = []
scores = []

with open('sample_data/scores.txt') as f:
    
    for line in f:
        print(line)    # iterable of strings, not subscriptable, requires seek 
        
    f.seek(0)
    
    lines = list(f)
    print(lines)             # list of strings 
    
    f.seek(0)
    
    reader = f.readlines()   # built in function
    print(reader)
    
    f.seek(0)    
    
    reader = csv.reader(f, delimiter=' ') 
    for line in reader:      # iterable of lists (ready stripped and split using delimiter), not subscriptable , requires seek 
        print(line)
        
    f.seek(0)    
    
    read = list(reader) # list of lists 
    print(read)
    
    f.seek(0)
    
    single_string = f.read()
    print(single_string)

    for line in f:
        print(line)
        
    f.seek(0)
        
    for line in lines:
        print(line)
        L = line.split()
        print(L)
        names.append(L[0])
        scores.append(L[1])
    print(names, scores)
        
```

```python
with open('sample_data/scores_.csv', 'w') as f:

    # loop through two lists 
    for n, s in zip(names, scores):
        f.write(n + ',' + str(s) + '\n')
```

```python
with open('sample_data/scores_.csv', 'w', newline='') as f:
    
    writer = csv.writer(f) 

    # loop through two lists 
    for n, s in zip(names, scores):
        writer.writerow([n, s])
```
